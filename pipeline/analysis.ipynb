```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
GG = pd.read_csv('CWQ Annotation - JP-Google (1).csv')
OURS = pd.read_csv('CWQ Annotation - JP (1).csv')
```

```python
GG_MP = list(GG['Meaning Preservation (1-5)'])[:42]
GG_F = list(GG['Fluency (1-5)'])[:42]
OURS_MP = list(OURS['Meaning Preservation (1-5)'])[:42]
OURS_F = list(OURS['Fluency (1-5)'])[:42]
```

```python
GG_F = [int(i) for i in GG_F]
OURS_F = [int(i) for i in OURS_F]
```

```python
step = 3
avg_gg_mp = []
avg_gg_f = []
avg_ours_mp = []
avg_ours_f = []
it = 0
while it<42:
    avg_gg_mp.append(np.mean(GG_MP[it:it+step]))
    avg_gg_f.append(np.mean(GG_F[it:it+step]))
    avg_ours_mp.append(np.mean(OURS_MP[it:it+step]))
    avg_ours_f.append(np.mean(OURS_F[it:it+step]))
    it+=step

```

```python
x = range(1,len(avg_gg_mp)+1)
x
```

```python
plt.style.use('seaborn-notebook')
plt.figure(figsize=(8,6))
plt.plot(x, avg_gg_mp, label = 'Google Translate', marker = 'o')
plt.plot(x, avg_ours_mp, label = 'Ours', marker = 'o')
plt.legend(loc=3, fontsize=12)
plt.xlabel('Complexity Level', fontsize=20)
plt.ylabel('Averaged Meaning Preservation', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
fig = plt.gcf()

fig.savefig('ComAvgMP.png', dpi=150)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, avg_gg_f, label = 'Google Translate', marker = 'o')
plt.plot(x, avg_ours_f, label = 'Ours', marker='o')
plt.xlabel('Complexity Level', fontsize=20)
plt.ylabel('Averaged Fluency', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.legend(loc=3, fontsize=12)
fig = plt.gcf()

fig.savefig('ComAvgF.png', dpi=150)
```
